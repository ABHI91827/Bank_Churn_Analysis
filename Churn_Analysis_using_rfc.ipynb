# STANDARDSCALER

```python
from sklearn.preprocessing import StandardScaler
sc=StandardScaler()
data[['CreditScore','Tenure','Balance','EstimatedSalary']]=sc.fit_transform(data[['CreditScore','Tenure','Balance','EstimatedSalary']])
```

# label encoding

```python
from sklearn.preprocessing import LabelEncoder
le=LabelEncoder()
data['Gender']=le.fit_transform(data['Gender'])
```

# one hot encoding


```python
dummies = pd.get_dummies(data[['Geography', 'Age']], drop_first=True)
data = pd.concat([data.drop(['Geography', 'Age'], axis=1), dummies], axis=1)
```

```python
x = data.drop(['RowNumber', 'CustomerId', 'Surname', 'Exited'], axis=1)
```

```python
y=data['Exited']
```

# Train and Test

```python
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.2,random_state=42)
```

# Model fitting

```python
model=RandomForestClassifier()
model.fit(x_train,y_train)
```

# prediction

```python
y_pred=model.predict(x_test)
```

# Accurancy

```python
print("confusion matrix",confusion_matrix(y_test,y_pred))
print("classification report",classification_report(y_test,y_pred))
print("accuracy score",accuracy_score(y_test,y_pred))
```
